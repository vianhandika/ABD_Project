```python
import fhirclient.models as m
import fhirclient.models.patient as p
import fhirclient.models.humanname as hn
import pandas as pd
import json
import numpy as np

```

```python
!unzip '/content/ABD_Project/Dataset/output/dataset3.zip'
data_patient = pd.read_csv('/content/dataset.csv')
data_patient

```

```python
# sorting by first name
data_patient.sort_values("DATE", ascending=False, inplace = True)
 
# dropping ALL duplicate values
# data_patient.drop_duplicates(subset ="PATIENT",keep = False, inplace = True)

data_patient=data_patient.drop_duplicates(subset='PATIENT')
data_patient
```

```python
data_patient.to_csv('dataset_patient.csv', index=False)
```

```python
patient_json = []
for index, row in data_patient.iterrows():
    #name
    p_name = m.humanname.HumanName({
        'use': str(row['MAIDEN']),
        'text' : str(row['FIRST'])+' '+str(row['LAST']),
        'family': str(row['LAST']),
        'given': [str(row['FIRST'])],
        'prefix':[str(row['PREFIX'])],
        'suffix':[str(row['SUFFIX'])]

    })
    #address
    p_address = m.address.Address({
        'text' : row['ADDRESS'],
        'city' : row['CITY'],
        'country' : row['COUNTY'],
        'state' : row['STATE'],
        'postalCode' : str(row['ZIP']),

    })
    
    #patient object
    patient = p.Patient({
        'id': row['Id.1'],
        'gender': row['GENDER'],
        'birthDate' : row['BIRTHDATE'],
        'deceasedBoolean' : False if pd.isna(row['DEATHDATE']) else True,

    })
    if pd.notna(row['DEATHDATE']):
        patient.deceasedDateTime = m.fhirdate.FHIRDate(row['DEATHDATE'])
    patient.address=[p_address]
    patient.name=[p_name]
    
    patient.as_json()
    patient_json.append(patient.as_json())
    # with open("Patient FHIR JSON/"+str(row['Id'])+"data.json", 'w', encoding='utf-8') as f:
    #     json.dump(patient.as_json(), f, ensure_ascii=False, indent=4)
# patient_json[:1]
```
