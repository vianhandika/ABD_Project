```python
data_panda = pd.read_csv('Dataset/dup_patients_preprocessing.csv')
data_panda
```

```python
data_spark = spark.read.csv('Dataset/dup_patients_preprocessing.csv', inferSchema = True, header = True)
data_spark.createOrReplaceTempView("table1")
data_spark.show()
```

```python
#Melakukan manipulasi data menggunakan pandas
%%time

for i in range(10):
    data_panda['CONDITION_HISTORY_TEMP'] = data_panda.CONDITION_HISTORY.replace(str(i), "")
    data_panda['count_'+str(i)] = data_panda.CONDITION_HISTORY.str.len() - data_panda.CONDITION_HISTORY_TEMP.str.len()
```

```python
data_panda
```

```python
#Melakukan manipulasi data menggunakan spark sql
%%time

for i in range(10):
    data_spark = data_spark.withColumn('CONDITION_HISTORY_TEMP', Func.regexp_replace('CONDITION_HISTORY',str(i),""))
    data_spark = data_spark.withColumn('count_'+str(i), Func.length("CONDITION_HISTORY") - Func.length("CONDITION_HISTORY_TEMP"))

data_spark.show()
```

Terbukti bahwa penggunaan pyspark dalam hal manipulasi data lebih cepat dari pada menggunakan pandas ketika diberi data yang banyak
